```python
import pandas as pd
```

```python
tmp = pd.read_excel('Config_31.xlsx')
```

```python
dct_tests = {}
for test in tmp.values:
    dct_tests[test[0]] = {tmp.columns[i]: test[i] for i in range(1, len(tmp.columns))}
```

```python
import json

json.dump(dct_tests, open('config.json', 'w', encoding='utf-8', errors='xmlcharrefreplace'))
```
